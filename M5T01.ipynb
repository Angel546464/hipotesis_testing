## EJERCICIO 1
* #### Calcula el p-valor y comprueba si se rechaza la hipótesis nul·la con un alfa de 5% para uno de los atributos.

```python
from scipy.stats import shapiro
stat, p = shapiro(ligas_uefa['Attendance'])
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
 print('Probablemente Gaussiana')
else:
 print('Probablemente no Gaussiana')
```

```python
import matplotlib.pyplot as plt
import scipy.stats as stats

stats.probplot(ligas_uefa['Attendance'], dist="norm", plot=plt)
plt.title("Q-Q Plot - Attendance")
plt.xlabel("Cuantiles teóricos")
plt.ylabel("Cuantiles observados")
plt.show()
```

Este gráfico compara la distribución real del atributo (en azul) sobre la línea roja que representa la distribución de los datos para una representación Gaussiana o normal. Hace la comparación a través de los quartiles reales (eje y) y los teóricos (eje x).

Se observa que los datos no están sobre la línea sobretodo al principio y al final.

## EJERCICIO 2
* #### Calcula el p-valor y comprueba si se rechaza la hipótesis nul·la con un alfa de 5% para dos de los atributos

Para este ejercicio selecciono las variables Goles a Favor y Puntos. Son dos variables que en principio no tendrían que tener una distribución normal, pero si una correlación entre ellas (a más goles a favor más posibilidad de sumar más puntos).

```python
stat1, p1 = shapiro(ligas_uefa['GF'])
stat2, p2 = shapiro(ligas_uefa['Pts'])

print('Goles a Favor: stat=%.3f, p=%.3f' % (stat1, p1))
if p1 > 0.05:
    print('Probablemente Gaussiana')
else:
    print('Probablemente no Gaussiana')

print('Puntos: stat=%.3f, p=%.3f' % (stat2, p2))
if p2 > 0.05:
    print('Probablemente Gaussiana')
else:
    print('Probablemente no Gaussiana')
```

```python
from scipy.stats import pearsonr
stat, p = pearsonr(ligas_uefa['GF'], ligas_uefa['Pts'])
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
 print('Probablemente independiente')
else:
 print('Probablemente dependiente')
```

## EJERCICIO 3
* #### Calcula el p-valor y comprueba si se rechaza la hipótesis nul·la con un alfa de 5% para tres de los atributos

Para este ejercicio usaré atributos estadísticos del conjunto de datos. En concreto trabajaré con xG (Expected Goals), xGA (Expected Goals Against) y xGD (Expected Goal Difference). 

Son métricas que evalúan las jugadas del partido y en base a muchos factores estiman si una jugada era más probable que acabara en gol o no. Esto se contrasta con los goles reales del partido y nos da estos atributos.

Primero exploro la normalidad de las tres variables.

```python
variable_xG = ligas_uefa['xG']
variable_xGA = ligas_uefa['xGA']
variable_xGD = ligas_uefa['xGD']

def shapiro_test(variable, title):
    stat, p = shapiro(variable)
    print(f'{title}: stat={stat:.3f}, p={p:.3f}')
    if p > 0.05:
        print('Probablemente Gaussiana')
    else:
        print('Probablemente no Gaussiana')

shapiro_test(variable_xG, 'xG')
shapiro_test(variable_xGA, 'xGA')
shapiro_test(variable_xGD, 'xGD')
```

En este caso vemos que los goles esperados (xG) no tienen una distribuicion normal, mientras que los goles en contra esperados (xGA) y la diferencia entre ambas métricas (xGD) si que tienen una distribución normal. Ahora paso a ver como se representan en una matriz de correlación. 

```python
corr_matrix = ligas_uefa[['xG', 'xGA', 'xGD']].corr()

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.show()
```

Nos indica una relación negativa moderada (-0.61) entre xG y xGA. 

También indica una relación negativa fuerte (-0.87) entre xGA y xGD. Esto indicaría que mientras más aumente una variable, más disminuirá la otra. 

El resultado 0.92 entre xG y xGD indica una fuerte correlación entre ambas variables. Mientras más goles esperados hay por partido el resultado para la diferencia entre los esperados recibidos y anotados también aumenta y viceversa.

Ahora muestro una visualizacion gráfica de estos resultados.

```python
sns.pairplot(ligas_uefa[['xG', 'xGA', 'xGD']])
```

Los histogramas en la diagonal muestran la normalidad de cada atributo. Vemos como xGA y xGD tienen una distribución más normalizada que xG.

En los gráficos de dispersión vemos mucha relación entyre xG y xGD así como entre xGA y xGD. En cambio vemos como entre xG y xGA el gráfico se muestra más disperso.
